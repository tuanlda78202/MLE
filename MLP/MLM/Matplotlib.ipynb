## `Pylot API` & `OO API` 

* More Information: https://bit.ly/2W28qxI

 ### `Pylot API` : quickly
 * allows us to interactively create simple plots.

```python
x = np.linspace(0,10,1000)

# plt.plot(x, y(x), color = "", linestyle = "", label = "" - ,must have legend)
plt.plot(x, np.sin(x), color = "red", linestyle = "-.", label = "sin(x)");
plt.plot(x, np.cos(x), color = "green", label = "cos(x)")

# show title
plt.title("Sin(x) & Cos(x)", fontsize = "20")

# show label of x-y axis
plt.xlabel("x", fontsize = "16")
plt.ylabel("f (x)", fontsize = "16")

# xmin, xmax, ymin, ymax = axis([xmin, xmax, ymin, ymax])
plt.axis([0,10,-1,1])
# moreover,we can use "tight" to fitting graph; "equal" (x axis = y axis)
plt.axis("equal")

# show label of plot
plt.legend()

# show graph
plt.show()
```

### `OO API` : advanced
* gives us more control when we create figures that contain multiple plots.

```python
# 1. Preparing data 
x = np.linspace(0,10,20)
y = np.linspace(20,80,20)

# 2. Setup Plot

# create a figure and a set of subplots(1,2,..)
# ax: object, plot with its methods
fig , ax = plt.subplots(figsize = (5,5)) 

# plot x y 
ax.plot(x,y)

# setup label
ax.set(title = "Graph Model",
      xlabel = 'x',
      ylabel = 'y (x)')

# 3. Show graph
plt.show()
```

# 2. Simple Plots 🏵

## 2.1. Line graph

* suitable for the purpose of showing up/down trends in time series and comparative data between groups in the category variable


```python
# creating data 
x = np.linspace(0,10,5)
```

```python
# Pyloy API 

# overview
plt.plot(x, x**2, marker = "o", linestyle = "-", label = 'x^2')

# labels
plt.xlabel("x", fontsize = 15)
plt.ylabel("y = x ^ 2 ", fontsize = 15)
plt.title("Parabol")

# show
plt.legend()
plt.show()
```

```python
# OO API 
fig, ax = plt.subplots(1, figsize = (5,5))

ax.plot(x,x**2)

ax.set(title = "Parabol",
      xlabel = "x-axis",
      ylabel = "y-axis")

# add text at an arbitrary location of the axes.
ax.text(6, 20, 
        "y = x ^ 2",
        bbox={},
        color='yellow',
        fontsize = 10)

plt.show()
```

## 2.2. Bar chart

* For comparing values between different groups

```python
# vertical bar
list_order = {"iphone 11" : 10,
             "iphone 12" :20,
             "mbp m1": 100,
             "mba m1": 200,
             "imac" :150}

fig, ax = plt.subplots()

ax.bar(list_order.keys(), list_order.values())

ax.set(title = "Apple Order",
      xlabel = 'Apple product',
      ylabel = 'Measured')

plt.show()
```

```python
# horizontal 

fig, ax = plt.subplots()
ax.barh(list(list_order.keys()), list(list_order.values()))

plt.show()
```

## 2.3. Pie chart

* Showing the market share of each segment or the percentage of each group

```python
# crawl data 
order_cpu = {"10105" : 100,
             "10300" :200,
             "11600": 70,
             "5900": 80,
             "5600" :50}
```

* By setting autopct to %.0f%%, we've chosen to format the percentages with 0 decimal places (only whole numbers), and added a % sign at the end. If we had omitted the surrounding %..% symbols, the strings wouldn't be formatted as percentages, but as literal values.

```python
fig, ax = plt.subplots(1)

# ax.pie(x, labels = labels, colors = colors, autopct='%.0f%%')
ax.pie(order_cpu.values(), labels = order_cpu.keys(), autopct='%.2f%%')

ax.set(title = "Market")

plt.show()
```

## 2.4. Scatter  

* The scatter plot is used to represent each point (x,y) in two-dimensional graphs. In machine learning, when we need to represent clusters or in situations where we want to find the distribution of data points along dimensions, we will often use a scatter plot.

```python
# Pylot API 

# size fig
plt.figure(figsize = (8,4))

# scatter
plt.scatter(df["Petal.Length"],df["Petal.Width"], color = "red", alpha = 0.4)

# label
plt.title("Average", fontsize = 18)
plt.xlabel("Petal Lengntsize = 15)
plt.ylabel("Petal Width", fontsize = 15)

# show
plt.show()
```

## 2.5. Histogram

* A graph representing the distribution of continuous variables by breaking them down into bins of equal length and counting the number of observations that fall into each bins.

```python
# preparing data 
np.random.seed(42)
student_height = np.random.normal(170, 20, 200)

# pylot API
plt.hist(student_height,bins = 30)

plt.show()
```

# 3. Subplots 🕊

* In reporting and analyzing many situations, we need to extract information from many related graphs. Therefore, we need to show many graphs on the same graph for easy comparison and conclusion.

* Matplotlib allows us to plot subplots on the same graph through a grid matrix.

```python
fig, ax= plt.subplots(nrows=2,
                      ncols =2,
                      figsize = (5,5))
# plot data to each axis 
ax[0,0].pie(order_cpu.values(), labels = order_cpu.keys(), autopct='%.2f%%')

ax[0,1].scatter(x, np.exp(x))

ax[1,0].barh(list(order_cpu.keys()), list((order_cpu).values()))

ax[1,1].hist(student_height, bins = 30)

# labels
fig.suptitle("Summary Charts", fontsize = 20)
plt.show()
```
