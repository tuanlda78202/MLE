## 3.3 Functions with data field


### 3.3.1. Min/Max, Sum, Mean and Median

```python
print(df['tax'].min(),df['tax'].max(),df['tax'].sum(),df['tax'].mean(),df['tax'].median(), sep = "\n")
```

### 3.3.2. Cut & Qcut

Effect: Seperating values of a continuos field to some segments.

**Cut Function** 

```python
bins = [0,20,40,50]
labels = ['low', 'normal', 'high']

df["medv_labels"] = pd.cut(df["medv"], bins = bins, labels = labels)

#show data(sort by medv)
df[df['medv_labels'] == 'high'].head()
```

**QCut (Quantile Cut)**

### 3.3.3. Apply

## 4.2. Pivotable

*statics in measurement vars about many dimesions*


```python
import numpy as np

pd.pivot_table(df, 
               columns = ['medv_labels', 'chas'], 
               index = ['rad'], 
               values = ['tax', 'rm'],
               aggfunc = {
                   'tax': np.sum,
                   'rm': np.mean
                   })
```
