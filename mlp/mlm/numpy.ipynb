## Sorting Array 
* np.sort() uses an quicksort algorithm.

```python
arr = np.array([8,4,2,13,1,5])

# if > 1dim, we can choose axis = 0 or 1 
np.sort(arr)
```

```python
# return index of element sorted
np.argsort(arr)

# index :  4,2,1,5,0,3
# element: 1,2,4,5,8,13
```

```python
# return index of max/min element with axis, default choose 1 arr in all matrix
# indices of max element - column
np.argmax(np.array([[1, 2],
              [3, 4],
              [5, 0],
              [2, 1]]), axis = 0)
```

```python
# indices of max element - row
np.argmax(np.array([[1, 2],
              [3, 4],
              [5, 0],
              [2, 1]]), axis = 1)
```

```python
# default the index is into the flattened array, otherwise along the specified axis {0;1}
np.argmax(np.array([[1, 2],
              [3, 4],
              [5, 0],
              [2, 1]]))
```

## 5.2. Vector 

* Vector is interpreted as a 1-dimensional array.

* Numpy does not refer to row vectors or column vectors, but only treats a vector as a one-dimensional array. If you want to **create a column vector** you have to treat it as a matrix with the second dimension equal to 1, and then you have to declare to numpy that it's a two-dimensional array

```python
# Column vector (a,1): [[1,2,3]].T or [[1],[2],[3]]
col_1 = np.array([[1],[2],[3]])
col_2 = np.array([[1,2,3]])

print(col_1,col_2, sep = '\n')
print(col_1.shape, col_2.T.shape)
```

```python
# shape = (4,) bcs define of tuple 1 ele. If 1 ele, define (a,)
arr_1 = np.array([1,2,3,4])
arr_2 = np.array([6,3,1,3])
print(arr_1.shape, arr_1.ndim, sep = "\n")
```

```python
# scalar: a x b = np.dot(a,b)
np.dot(arr_1,arr_2)
```

```python
# direction
arr_1 * arr_2
```
