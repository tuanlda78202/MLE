## 4.1. df.groupby()

* groupby is a command that allows you to apply functions on **measurements** based on grouping data by **dimensions**.

* measurements are variables that can be added, subtracted, multiplied, divided, and measured, while dimensions are variables used to group data

**Example**: Suppose we have a list of food orders in 1 day (hundreds, thousands of customers). The problem is which dish is ordered the most?



* Using groupby() we will cluster the groups into groups one by one, then use the .sum() function:



* Next, we can **.sort_value(ascending = False)**

```python
# df.groupby("dimesion")[measurements]

df.groupby("rad")["tax"].apply(print)
```

```python
# moreover, we can def function to apply(lambda)

def square(x):
  return x ** 2

df.groupby("rad")["tax"].apply(lambda x: square(x))
```

## 4.2. Pivotable

>statics in measurement vars about many dimesions

* columns: List the dimensions of the column that we need to calculate.
* index: List of dimensions by row that we need to statistics.
* values: List of measurements we use to calculate.
* aggfunc: Specify the function we will use to transform the measurement. In this example we apply the function np.sum to the math set of measurements.

```python
import numpy as np

pd.pivot_table(df, 
               columns = ['medv_labels', 'chas'], 
               index = ['rad'], 
               values = ['tax', 'rm'],
               aggfunc = {
                   'tax': np.sum,
                   'rm': np.mean
                   })
```
