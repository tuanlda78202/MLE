```python
import numpy as np
```

```python
def grad(x):
    return 2*x + 5*np.cos(x)

def cost(x):
    return x**2 + 5*np.sin(x)

def GD(x0, alpha):
    # create list to access end element of list to calculating
    list_x = [x0]
    # for loop to update x0 w learning rate, maybe 1e3,1e4
    for i in range(100):
        # formula 
        x_new = list_x[-1] - alpha * grad(list_x[-1])
        # if grad ~ 0, then can inconcluding ~ df=0
        if abs(grad(x_new)) < 1e-3: #1e-3 = 10^-3
            break
        list_x.append(x_new)
    return (list_x, i)
```

## 1. Different starting points

```python
# sol: list_x[-1], after i interations.
print(GD(-5,.1))
```



With different starting points, the algorithm finds nearly the same solution, although with different convergence rates.

In Figure 12.3 corresponding to x0 = 5, the path of the solution contains an area of rather small derivatives near the point of coordinate 2.5. This causes the algorithm to hang around here for quite some time. Once this point is passed, everything goes very well. Points that are not minima but have derivatives close to zero are very easy to cause a trapped solution here because the small derivative makes it not change much in the next iteration.



## 2. Different Learning Rate


In fact, when the calculation becomes complicated, too low a learning rate will affect the speed of the algorithm greatly, even never reaching the destination.



The choice of learning rate is very important. This depends a lot on the problem and it takes several experiments to choose the best value. In addition, depending on some problems, GD can work more efficiently by choosing the right learning rate or choosing a different learning rate in each iteration, usually gradually.
