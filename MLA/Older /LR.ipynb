```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[147,150,153,155,158,160,163,165,168,170,173,175,178,180,183]]).T
y = np.array([49,50,51,52,54,56,58,59,60,62,63,64,66,67,68])

plt.plot(X,y, marker = 'o', color = 'green', linestyle = '-', label = 'line')

plt.axis('tight')
plt.axis([150,180,50,80])

plt.title('Line Graph')
plt.xlabel('X', fontsize = 15)
plt.ylabel('y', fontsize = 15)

plt.legend()
plt.show()
```

```python
matrix_ones = np.ones((X.shape[0],1), dtype = int)

X_plus = np.concatenate((matrix_ones,X), axis = 1)
X_plus
```

```python
inverse = np.linalg.pinv(np.dot(X_plus,X_plus.T))
pre_result = np.dot(inverse, X_plus)
result = np.dot(pre_result.T,y)

print(result)
print(result[0] + result[1] * 155)
```
